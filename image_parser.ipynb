```python
import os
from PIL import Image
import numpy as np
```

```python
os.makedirs('lowercase_characters', exist_ok=True)
img = Image.open('handwriting_samples/lowercase.jpg')

width, height = img.size
rows, cols = 26, 30

cell_w = width // cols
cell_h = height // rows

for i in range(rows):
    char = chr(ord('a') + i)
    os.makedirs(f'lowercase_characters/{char}', exist_ok=True)

    for j in range(cols):
        left = j * cell_w
        right = (j+1) * cell_w
        top = i * cell_h
        bottom = (i+1) * cell_h

        char_img = img.crop((left, top, right, bottom))
        arr = np.array(char_img)

        arr = arr < 128
        used_cols = np.where(arr.any(axis=0))[0]

        crop_left = used_cols[0]
        crop_right = used_cols[-1] + 1

        char_img = char_img.crop((crop_left, 0, crop_right, cell_h))
        char_img.save(f'lowercase_characters/{char}/{j}.jpg')


```

```python
os.makedirs('punctuation', exist_ok=True)
img = Image.open('handwriting_samples/punctuation.jpg')

width, height = img.size
rows, cols = 26, 30

used_rows = 8

punctuations = ".,';:?!\""


cell_w = width // cols
cell_h = height // rows

for i in range(rows):
    if i >= used_rows:
        break
    char = punctuations[i]
    os.makedirs(f'punctuation/{char}', exist_ok=True)

    for j in range(cols):
        left = j * cell_w
        right = (j+1) * cell_w
        top = i * cell_h
        bottom = (i+1) * cell_h

        char_img = img.crop((left, top, right, bottom))
        arr = np.array(char_img)

        arr = arr < 128
        used_cols = np.where(arr.any(axis=0))[0]

        crop_left = used_cols[0]
        crop_right = used_cols[-1] + 1

        char_img = char_img.crop((crop_left, 0, crop_right, cell_h))
        char_img.save(f'punctuation/{char}/{j}.jpg')


```
